# Code for Web Scraping - Selenium - Bienici.com

```python
#importing packages

from selenium import webdriver
import pandas as pd 
import time
import numpy as np
import os
from os.path import dirname, join
```

```python
#Creating a list of webpages to scrape data from

token1 = 'https://www.bienici.com/recherche/achat/lille-59000?page='
token2 = 'https://www.bienici.com/recherche/location/lille-59000?page='
token3 = 'https://www.bienici.com/recherche/achat/lille-59000?neuf=oui&page='


pages = []
for i in range(1,12):
    pages.append(token1+str(i))
for i in range(1,12):
    pages.append(token2+str(i))
for i in range(1,3):
    pages.append(token3+str(i))
    
pages
```

```python
#Empty lists to which data will appended

names = []
addresses = []
prices = []
types = []
```

```python
#Initialize driver
driver = webdriver.Firefox()
driver.implicitly_wait(60)
```

```python
'''
Selenium creates a list of all the classes it finds in a webpage, and stores it a variable. Each list 
element (class) contains a .text which will have the text data from the webpage. 
The variables 'name' , 'address' and 'price' are the list of class elements in the webpage which contain the data.
We append this .text to an empty list which we use to create the dataframe.
'''
for i in range(0,len(pages)):
    driver.get(pages[i])
    driver.implicitly_wait(60)
    #good practice 
    
    driver.find_element_by_xpath('//*[@id="searchSideHeader"]/div/div[2]/div/div/div/div[2]').click()
    #expand the "searchSideHeader" class to find elements in it
    
    
    name = driver.find_elements_by_class_name('descriptionTitle')
    address = driver.find_elements_by_class_name('descriptionTitleAddress')
    price = driver.find_elements_by_class_name('thePrice')
    #create a list of elements with the specified name
    #e.g., 'name' is a list of all the elements in webpage with class name 'descriptionTitle'
    
    

    for a in name:
        names.append(a.text)
    
    for b in address:
        addresses.append(b.text)
        
    for c in price:
        if c.text != "":
            prices.append(c.text)
    
    #append the text from each element of elements list to new list, i.e, the data list
    #e.g., appending text (data) of elements in 'name' list to 'names' list 
            
        
    
'''
BLOGS I REFERRED TO:

https://medium.com/ml-book/web-scraping-using-selenium-python-3be7b8762747
https://lewiskori.com/blog/beginner-s-guide-to-web-scraping-with-python-s-selenium/

'''
    
```

```python
driver.quit()
```

```python
prices.append('') #last element did not have price data 

data = {'category' : names,'address' : addresses, 'price' : prices}
#data dictionary
```
